# Máscaras con OpenCV

### Importar paquetes y librerias

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

### Función para mostrar una imagen con imshow() de pyplot

```python
def mostrar_imagen(title, imagen):
  imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
  plt.imshow(imagen)
  plt.title(title)
  plt.grid(False)
  plt.show()
```

### Máscaras

```python
args = {
    "imagen" :"autisha.jpg"
}
```

```python
# cargar la imagen original y mostrarla
imagen = cv2.imread(args["imagen"])
mostrar_imagen("Original",imagen)
```

La máscara tiene el mismo tamaño de la imagen, pero solamente tiene dos valores de pixel, 0 y 255. Los pixeles de la máscara con valor 0 son ignorados en la imagen original, mientras que los pixeles de la máscara con valores 255 se conservan. 

```python
imagen.shape[:3]
```

**Máscara Rectangular**

```python
mascara = np.zeros(imagen.shape[:2],dtype="uint8")
cv2.rectangle(mascara,(0,0),(150,150),255,-1)
mostrar_imagen("Máscara Rectangular", mascara)
```

```python
conMascara = cv2.bitwise_and(imagen,imagen,mask=mascara)
mostrar_imagen("Máscara aplicada a la imagen",conMascara)
```

**Máscara Circular**

Esta máscara circular tiene un radio igual a 50 pixeles

```python
mascara = np.zeros(imagen.shape[:2],dtype="uint8")
cv2.circle(mascara,(50,50),50,255,-1)
mostrar_imagen("Máscara Circular", mascara)
```

```python
conMascara = cv2.bitwise_and(imagen,imagen,mask=mascara)
mostrar_imagen("Máscara aplicada a la imagen",conMascara)
```
