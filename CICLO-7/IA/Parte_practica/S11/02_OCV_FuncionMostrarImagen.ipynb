```python
from matplotlib import pyplot as plt
import cv2
```

```python
def mostrar_imagen(titulo,imagen):
  imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
  plt.imshow(imagen)
  plt.title(titulo)
  plt.grid(False)
  plt.show
```

```python
imagen = cv2.imread("valve.jpg")
mostrar_imagen("Open Your Mind",imagen)
```
