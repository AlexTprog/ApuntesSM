# Deteccion de Rostros

Importando librerias

```python
from detector.detector_de_rostros import DetectorRostro
import argparse
import cv2
from matplotlib.pyplot import imshow
```

Definimos argumentos

```python
args = {
    "imagen":"/home/alex/Documentos/ApuntesSM/CICLO-VII/IA/Parte_practica/semana14/Pro02DeteccionRostros/imagenes/prueba00.jpeg",
    "imagen2":"/home/alex/Documentos/ApuntesSM/CICLO-VII/IA/Parte_practica/semana14/Pro02DeteccionRostros/imagenes/prueba01.jpeg",
    "rostro":"/home/alex/Documentos/ApuntesSM/CICLO-VII/IA/Parte_practica/semana14/Pro02DeteccionRostros/cascadas/haarcascade_frontalface_default.xml"
}
```

```python
# cargar la imagen y convertirla a escala de grises
imagen = cv2.imread(args["imagen"])
gris = cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)
```

```python
# encontrar los rostros en la imagen
dr = DetectorRostro(args["rostro"])
rectRostros = dr.detectar(gris,factorEscala=1.1, vecMin=5, tamMin=(30, 30))
print("Se encontró {} rostro(s)".format(len(rectRostros)))
```

```python
# dibujar los rectángulos sobre los rostros
for(x , y, w, h) in rectRostros:
    cv2.rectangle(imagen,(x,y),(x+w,y+h),(0,255,0),2)
imshow(imagen)
```
